```python
df = pd.read_csv("https://raw.githubusercontent.com/DanielRogate/datasetdatmin2/main/posttest2-_1_.csv") # potongan code ini digunakan untuk membaca file excel dataset.xlsx dengan method read_excel dengan sheet DatasetBaru - Sheet 1
df # potongan cide ini digunakan untuk memperlihat Data Tabular item pada variavel dfa
```

```python
from itertools import permutations # potongan code ini digunakan sebagai library dalam membuat kombinasi barang
rules = list(permutations(df.columns, 2)) # potongan code ini digunakan untuk menyimpan seluruh data kemungkinan kombinasi barang
rules # potongan code ini digunakan untuk menampilkan keseluruhan hasil dari setiap kombinasi
```

```python
rule_baru = [] # potongan code ini digunakan untuk menyimpan data dalam bentuk list
for rule in rules: # potongan code ini digunakan untuk mengulangi pembacaan dari dataset
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0: # potongan code ini digunakan untuk menentukan apakah kombinasi tersebut ada dalam data atau sebaliknya
    rule_baru.append(rule) # potongan code ini digunakan untuk menyimpan hasil dari pengecekan dalam bentuk tuple ke dalam list
rule_baru # potongan code ini digunakan untuk menampilkan hasil kombinasi dalam data
```

```python
hitung = {} # potongan code ini digunakan untuk menyimpan data ke dalam bentuk dictionary
for rule in rule_baru:  # potongan code ini digunakan sebagai perulangan untuk mengulangi pembacaan pada dataset
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1) # potongan code ini digunakan untuk menentukan berapa banyak jumlah kemunculan setiap kombinasi lalu menyimpan datanya ke dalam sebuah variabel
      ])
  hitung[str(rule)] = banyak_data # potongan code ini digunakan untuk menambahkan hasil penentuan ke dalam dictionary
hitung # potongan code ini digunakan untuk menampilkan hasil kombinasi dalam data
```

```python
len(df) # potongan code ini digunakan untuk melihat jumlah transaksi
```

```python
for item,ab in hitung.items():
  print(f"{item} support = {ab/len(df)*100}%")
```

```python
for item,ab in hitung.items():
  i = eval(item)
  print(f"{i[0]}->{i[1]}  Confidence = {ab/df[i[0]].sum()*100}%")
```

## 1 Membuat tabel jarak berdasarkan data baru yang ingin ditambah

```python
import numpy as np

# potongan code ini digunakan untuk membuat dataframe dari data yang baru
new_data = pd.DataFrame({'diagonal'    : [172.25],
                         'height_left' : [103.94],
                         'height_right': [103.68],
                         'margin_low'  : [4.14],
                         'margin_up'   : [3.25],
                         'length'      : [112.48]})

# potongan code ini digunakan untuk menghitung jarak antara setiap baris dataframe utama dengan dataframe data baru
distances = []
for i in range(len(df2)):
    row = df2.iloc[i]
    distance = np.sqrt((row['diagonal'] - new_data['diagonal']) ** 2 +
                       (row['height_left'] - new_data['height_left']) ** 2 +
                       (row['height_right'] - new_data['height_right']) ** 2 +
                       (row['margin_low'] - new_data['margin_low']) ** 2 +
                       (row['margin_up'] - new_data['margin_up']) ** 2 +
                       (row['length'] - new_data['length']) ** 2)
    distances.append(distance)

# potongan code ini digunakan untuk menambahkan kolom distance ke dataframe utama
df2['distance'] = distances

# potongan code ini digunakan untuk menampilkan tabel jarak
print('Tabel Jarak : ')
print(df2['distance'])
```
